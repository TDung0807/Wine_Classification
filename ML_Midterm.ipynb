## (1) Phương pháp giải quyết bài toán học máy bằng Classification

1. Xác định vấn đề: Trong mọi bài toánn học máy đều cần bắt đầu bằng việc xác định vấn đề. Phải nắm rõ được mục tiêu cần phải đạt được điều gì với mô hình học máy. Ví dụ, trong bài toán phân lớp (classification) cần phân lớp email rác, tệp tin có mã độc hay không,…
2. Thu thập và tiền xử lý dữ liệu: Dữ liệu đóng vai trò thiết yếu trong việc quyết định kết quả của một mô hình học máy. Thu thập dữ liệu liên quan đến bài toán từ các nguồn khác nhau. Dữ liệu trong bài toán phân lớp có thể bao gồm các đặc trưng (features) và nhãn (labels) tương ứng với mỗi mẫu dữ liệu. Tiền xử lý dữ liệu bao gồm việc giải quyết các vấn đề như loại bỏ các dữ liệu bị thiếu, chuẩn hóa dữ liệu, xử lý các điểm ngoại lệ, loại bỏ nhiễu và các bước khác nếu cần tùy thuộc vào bài toán cụ thể. Bước này cũng bao gồm việc chia dữ liệu thành các tập huấn luyện (để huấn luyện mô hình) và tập kiểm tra (để đánh giá hiệu suất của mô hình đã huấn luyện). Thông thường, tỉ lệ giữa tập huấn luyện và tập kiểm tra là 80%-20% hoặc 70%-30% tùy thuộc vào bài toán và kích thước của tập dữ liệu.
3. Lựa chọn và xây dựng mô hình: Chọn lựa một mô hình học máy phù hợp với yêu cầu của bài toán. Đối với phân loại, các mô hình phổ biến bao gồm Hồi quy Logistic, Cây Quyết định, Rừng Ngẫu nhiên, Máy Vectơ Hỗ trợ (SVM), Mạng Nơ-ron, và nhiều hơn nữa. Trong khi đó, các mô hình cho bài toán hồi quy bao gồm Hồi quy Tuyến tính, Cây Quyết định, Rừng Ngẫu nhiên và Mạng Nơ-ron.
4. Huấn luyện mô hình: Trong quá trình này, mô hình “học” từ tập huấn luyện để tìm ra các tham số tối ưu, thường qua việc tối ưu hóa hàm mất mát (loss function) bằng cách sử dụng các phương pháp như Gradient Descent hoặc các biến thể của nó. Mục tiêu là điều chỉnh các tham số của mô hình để dự đoán của nó càng gần với thực tế càng tốt.
5. Đánh giá mô hình: Mô hình sau khi được huấn luyện cần được kiểm tra trên tập dữ liệu kiểm tra để đánh giá hiệu suất. Các độ đo thông thường trong bài toán phân lớp bao gồm độ chính xác (accuracy), độ phủ (recall),  độ phân loại (precision), F1-score và ma trận nhầm lẫn (confusion matrix). Độ đo này giúp đánh giá khả năng dự đoán của mô hình và đánh giá hiệu suất của nó trên các lớp nhãn khác nhau.
6. Tinh chỉnh và tối ưu hóa: Nếu mô hình không đạt được kỳ vọng, ta có thể tinh chỉnh và tối ưu hóa mô hình bằng cách điều chỉnh các tham số, thử nghiệm các mô hình khác nhau, sử dụng kỹ thuật tăng cường dữ liệu (data augmentation), hay áp dụng các kỷ thuật tinh chỉnh siêu tham số (hyperparameter tuning) để cải thiện hiệu suất của mô hình.
7. Triển khai: Sau khi mô hình đã được huấn luyện và đạt được hiệu quả đáng chấp nhận, nó có thể được triển khai trong môi trường thực tế. Điều này thường liên quan đến việc tích hợp nó vào trong một hệ thống hoặc một ứng dụng.
8. Giám sát và bảo trì: Các mô hình học máy cần được giám sát liên tục để đảm bảo rằng chúng hoạt động đúng như mong đợi theo thời gian. Khi dữ liệu phát triển, mô hình có thể cần được huấn luyện lại để duy trì độ chính xác.

### Build Models

```python
class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X_train, y_train):
        # Lưu trữ dữ liệu huấn luyện
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        predictions = []
        for sample in X_test:
            distances = []
            # Tính khoảng cách từ mỗi điểm dữ liệu trong tập kiểm tra đến điểm dữ liệu kiểm tra
            for train_sample, label in zip(self.X_train, self.y_train):
                distance = np.linalg.norm(sample - train_sample)  # Tính khoảng cách Euclid
                distances.append((distance, label))
            distances.sort(key=lambda x: x[0])  # Sắp xếp theo khoảng cách
            k_nearest = distances[:self.k]  # Chọn K láng giềng gần nhất
            k_nearest_labels = [label for _, label in k_nearest]  # Lấy nhãn của các láng giềng
            # dự đoán nhãn cho điểm dữ liệu mới
            prediction = max(set(k_nearest_labels), key=k_nearest_labels.count)
            predictions.append(prediction)
        return predictions


```

```python
class GaussianNB:
    def fit(self, X, y):
        self.X = X
        self.y = y
        self.classes = np.unique(y)
        self.parameters = []

        for c in self.classes:
            X_c = X[y == c]
            mean = np.mean(X_c, axis=0)
            var = np.var(X_c, axis=0)
            parameters = {"mean": mean, "var": var}
            self.parameters.append(parameters)

    def _calculate_likelihood(self, mean, var, x):
        # Hàm tính likelihood theo phân phối Gaussian
        exponent = np.exp(-(x - mean) ** 2 / (2 * var))
        return (1 / (np.sqrt(2 * np.pi * var))) * exponent

    def _calculate_prior(self, c):
        # Tính xác suất tiên nghiệm của lớp c
        return np.mean(self.y == c)

    def predict(self, X):
        # Dự đoán nhãn cho dữ liệu mới
        posteriors = []
        for i, c in enumerate(self.classes):
            prior = self._calculate_prior(c)
            posterior = np.prod(self._calculate_likelihood(self.parameters[i]["mean"],
                                                           self.parameters[i]["var"], X), axis=1) * prior
            posteriors.append(posterior)
        posteriors = np.array(posteriors).T
        return self.classes[np.argmax(posteriors, axis=1)]
```

```python
# build model
knn_custom = KNN(k=10)
```

```python
# naive bayes model (Gaussian)
gauss_model = GaussianNB()
```

```python
# logistic model
lr_model = LogisticRegression(C= 0.5, solver='liblinear',tol=0.5)
```

```python
# naive bayes model (MultinormalNB)
multiNB_model = MultinomialNB(fit_prior=False)
```

## (3) Nâng độ chính xác của các mô hình nêu ở câu (2)

### Tinh chỉnh Siêu tham số

Tinh chỉnh siêu tham số là quá trình tìm kiếm và chọn lựa các giá trị tối ưu cho các tham số quan trọng của một mô hình học máy. Quá trình này thường được thực hiện thông qua việc sử dụng các phương pháp như Grid Search và Randomized Search.

**Grid Search:**
Grid Search là phương pháp duyệt qua tất cả các tổ hợp có thể của các giá trị tham số được xác định trước và đánh giá hiệu suất của mô hình cho mỗi tổ hợp. Quá trình này tốn nhiều thời gian tính toán nhưng đảm bảo tìm ra giá trị tối ưu.

**Randomized Search:**
Randomized Search lựa chọn ngẫu nhiên một số lượng giới hạn của các tổ hợp tham số để đánh giá hiệu suất của mô hình. Đây là một phương pháp hiệu quả hơn về mặt tính toán so với Grid Search, nhưng có thể không đảm bảo tìm ra giá trị tối ưu nhất.

#### LogisticRegesstion

```python
param_dist = {'C': uniform(loc=0, scale=1),
              'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
              'tol': uniform(loc=0, scale=0.01)}

logistic_model = LogisticRegression()

random_search_logistic = RandomizedSearchCV(estimator=logistic_model,
                                            param_distributions=param_dist,
                                            n_iter=100, cv=5, random_state=42)
random_search_logistic.fit(X_train, y_train)

best_params_logistic = random_search_logistic.best_params_
print("Best parameters:", best_params_logistic)
print("Best score:", random_search_logistic.best_score_)
```

#### MultinomialNB

```python
param_dist = {
    'alpha': uniform(loc=0, scale=1),
    'fit_prior': [True, False],
}

multiNB_model = MultinomialNB()

random_search_multiNB = RandomizedSearchCV(estimator=multiNB_model, param_distributions=param_dist, n_iter=100, cv=5, random_state=42)

random_search_multiNB.fit(X_train, y_train)

best_params_multiNB = random_search_multiNB.best_params_
print("Best parameters:", best_params_multiNB)
print("Best score:", random_search_multiNB.best_score_)
```

### Đánh giá sau khi tối ưu

```python
lr_model = LogisticRegression(**best_params_logistic)
lr_model.fit(X_train, y_train)
# predict
y_predict = lr_model.predict(X_test)
# print(y_predict)
# Đánh giá hiệu suất của mô hình
print("Mô hình trước khi tinh chỉnh:")
print(lr_report)
accuracy_lr = accuracy_score(y_test, y_predict)
print("Accuracy:", accuracy_lr)
report = classification_report(y_test, y_predict)
print("Mô hình sau khi tinh chỉnh:")
print(report)
```

```python
multiNB_model = MultinomialNB(**best_params_multiNB)
multiNB_model.fit(X_train, y_train)

# Dự đoán nhãn cho tập kiểm tra
y_predict = multiNB_model.predict(X_test)

# Đánh giá hiệu suất của mô hình
print("Mô hình trước khi tinh chỉnh:")
print(multiNB_report)
print("Mô hình sau khi tinh chỉnh:")
accuracy_multiNB = accuracy_score(y_test, y_predict)
print("Accuracy:", accuracy_multiNB)

# In ra báo cáo đánh giá chi tiết
report = classification_report(y_test, y_predict)
print(report)
```
